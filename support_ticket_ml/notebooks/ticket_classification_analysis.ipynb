# Customer Support Ticket Classification - Exploratory Data Analysis
This notebook explores the support ticket dataset, visualizes category distributions, and performs an analysis of commonly used words in ticket texts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
```

```python
# Load processed dataset
df = pd.read_csv('../data/processed_tickets.csv')
display(df.head())
```

```python
# Plot Distribution of Ticket Categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', order=df['category'].value_counts().index, palette='Set2')
plt.title('Distribution of Support Ticket Categories')
plt.xticks(rotation=45)
plt.tight_layout()

import os
os.makedirs('../visualizations', exist_ok=True)
plt.savefig('../visualizations/ticket_category_distribution.png')
plt.show()
```

```python
# Plot Distribution of Priority Levels
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='priority', order=['High', 'Medium', 'Low'], palette='viridis')
plt.title('Distribution of Priority Levels')
plt.tight_layout()
plt.savefig('../visualizations/priority_distribution.png')
plt.show()
```

```python
# Most Common Words in Tickets
words = ' '.join(df['processed_text'].dropna()).split()
word_counts = Counter(words)
common_words = word_counts.most_common(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=[w[1] for w in common_words], y=[w[0] for w in common_words], palette='mako')
plt.title('Top 20 Most Common Words in Support Tickets')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.tight_layout()
plt.show()
```
